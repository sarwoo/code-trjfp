```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_file_names = ['Intercept_2021.csv', 'Intercept-KPI-Historic-21.csv']
data_directory = './data/'
```

Concat and read in all raw data files and set the intercept date to aa datetime

```python
data_files = [f'{data_directory}{file}' for file in data_file_names]
df = pd.concat(map(pd.read_csv, data_files), ignore_index=True)

df["Date Intercepted"] = pd.to_datetime(df["Date Intercepted"], dayfirst=True)
```

```python
filter_month = df['Date Intercepted'].dt.month == 10
print(f"October: {df.loc[filter_month]['Weight (in grams)'].sum():,}g")
```

```python
data = df.loc[filter_month, ['Weight (in grams)', "Product"]]
```

```python
plt.style.use("fast")
data_x = df['Product']
data_y = df['Weight (in grams)'] / 1_000_000
plt.barh(data_x, data_y)
plt.tight_layout
plt.show()
```
